**BASIC WORDCLOUD IMPLEMENTATION**

```
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a list of words
text=(" I am a cool guy . Don't know where I've been where would I be. I hope I travel and explore enough because the more you explore the more you know about yourself. It is well know that you test yourself, when curious situations arise. Only in these cases that you come to kno wof yourself that you didn't know.Explore and don't be afraid. This is the time ot make mistakes and RISE high.")

#Create the wordcloud object
wordcloud = WordCloud(width=480, height=480, margin=0).generate(text)

#Display the the generated image:

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

**WordCloud with a SPECIFIC SHAPE**

```
from wordcloud import WordCloud
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import cv2

text=(" I am a cool guy . Don't know where I've been where would I be. I hope I travel and explore enough because the more you explore the more you know about yourself. It is well know that you test yourself, when curious situations arise. Only in these cases that you come to kno wof yourself that you didn't know.Explore and don't be afraid. This is the time ot make mistakes and RISE high.")
#Load the image for mask
script=open("wordcloud_batman.txt").read()
wave_mask = np.array(Image.open("bike.jpeg"))

#Make the figure
wc = WordCloud(mask=wave_mask,
               width=2000,
               height=1500,
               contour_width=1,
               scale=80, 
               contour_color="White",
               colormap="gist_gray", 
               background_color="Skyblue",
               max_font_size=6, 
               max_words=88000,
               repeat=True
               )
wc.generate(script)

plt.figure(figsize=(8,12), facecolor='k')
plt.imshow(wc, interpolation="hermite")
plt.axis("off")
plt.figure(figsize=(4,12), facecolor='k')


plt.imshow(wave_mask, cmap = plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.tight_layout(pad=0)
plt.show()

```

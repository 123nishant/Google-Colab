# To create a recommendation , we need vector matrix of each movies

#tf-idf stands for Term frequency-inverse document frequency. 
#The tf-idf weight is a weight often used in information retrieval and text mining. 
#Variations of the tf-idf weighting scheme are often used by search engines in scoring and
#ranking a document's relevance given a query.

# Creates a Document Vector Matrix for sentences

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(stop_words='english',
                     min_df=3,
                     max_features=None,
                     strip_accents='unicode',
                     analyzer='word',
                    token_pattern=r'\w{1,}',
                     ngram_range=(1,3))

# ngram_range will take combination of 1 to 3 different kind of words
# Stop words is used to ignore words like 'the','a','is', etc types of wordss
# strip_accents , analyzer and token_pattern is used to remove unnecessary letters like
# '!',',','.' i.e. special symbols

# fill the null values
movies['overview']=movies['overview'].fillna('')
```

```python
tfv_matrix = tfv.fit_transform(movies['overview'])

#Learn vocabulary and idf, return term-document matrix.

#This is equivalent to fit followed by transform, but more efficiently implemented.
```

```python
tfv_matrix
```

```python
tfv_matrix.shape
```

 Use sigmoid kernel as sigmoid is between 0 and 1

# if sigmoid of sentence 1 and 2 is 0.5 ,
# then movie 1 is hald similar to movie 2

```python
from sklearn.metrics.pairwise import sigmoid_kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)

#computing a sigmoid if 2 similar matrices so that below dataframe is generated
# a dataframe of similarity scores with each movie is generated and many would have null values
# because of null ismilarity

```

```python
sig.shape
```

```python
# Giving indices to movies
indices = pd.Series(movies.index, index = movies['original_title']).drop_duplicates()
```

```python
indices
```

```python
#index of the maze runner
indices['The Maze Runner']
```

# Function for Content Based Recommendation System


```python
def recommend(title, sig=sig):
    
    # Here title is in text and we will get the respective index assigned to it
    index = indices[title]
    
    ####GOLDEN STEP:
    #lets calculate the sigmoid scores of this movie with every other movie
    sigmoid_scores = list(enumerate(sig[index]))
    
    # Sort it descending so that top correlated can be listed on top
    sigmoid_scores = sorted(sigmoid_scores,key = lambda x:x[1], reverse = True)
    # We will take only top 5 recommendations
    
    sigmoid_scores = sigmoid_scores[1:6]
    
    movie_indices = [i[0] for i in sigmoid_scores]
    
    return movies['original_title'].iloc[movie_indices]
```

```python
print('Top recommendations for this movie is : ')
recommend('The Dark Knight')
```

# Method 2 for Content Based Filtering

For Vectoorizing we will use **Count Vectorizer**
And for Similarity scoring we will use **Cosine Similarity**

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# dataset
movies.head()
```

```python
cv = CountVectorizer(stop_words='english')
cv_matrix = cv.fit_transform(movies['overview'])
```

```python
cv_matrix.shape
```

```python
cv_matrix
```

```python
cos_similarity = cosine_similarity(cv_matrix)
```

```python
cos_similarity # Shows the similarity of one movie with other movies
```
